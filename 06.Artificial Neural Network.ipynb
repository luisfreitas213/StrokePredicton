```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense
from keras.utils.vis_utils import plot_model
```

```python
X_train = pd.read_csv('X_train.csv')
y_train = pd.read_csv('y_train.csv')
X_train_oversampled = pd.read_csv('X_train_oversampled.csv')
y_train_oversampled = pd.read_csv('y_train_oversampled.csv')
X_test = pd.read_csv('X_test.csv')
y_test = pd.read_csv('y_test.csv')
```

```python
#Feature Scalling 
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(X_train_oversampled)
x_test = sc.transform(X_test)

print(x_train)
print(x_test)
```

```python
#Initialize ANN 
ann = tf.keras.models.Sequential()

#Adding the input layer and first 
ann.add(tf.keras.layers.Dense(units = 20, activation='relu')) # Camada Dense (Conectada a 100 %); 6 Neuronios na camada oculta e funçaõ de ativação relu

#Adding The Second Hidden Layer
ann.add(tf.keras.layers.Dense(units = 20, activation='relu')) # Camada Dense (Conectada a 100 %); 6 Neuronios na camada oculta e funçaõ de ativação relu

#Adding The Output Layer 
ann.add(tf.keras.layers.Dense(units = 1, activation='sigmoid')) # 1 neurónio de saída porque é binário; funcao sigmoid para ter a probabilidade do resultado

#Compiling the Ann 
ann.compile(optimizer = 'sgd', loss = 'binary_crossentropy', metrics = ['AUC']) # optimezer é adam porque faz a descida do gradiente estocastico
#print(ann.summary())
```

```python
#Compiling the Ann 
#Training the ANN on the Training Set
ann.fit(x_train, y_train_oversampled, batch_size = 63, epochs = 20)
```

```python
plot_model(ann, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

```python
print(ann.summary())
```

```python
y_pred = ann.predict(x_test)
y_pred = y_pred > 0.5
print (y_pred)
print(y_test)
y_test = y_test.values
predictons = np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1)
for i in predictons:
    print(i)
```

```python
"""Confusion Matrix"""
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
cm = confusion_matrix(y_test, y_pred)
acc = accuracy_score(y_test, y_pred)
ps = precision_score(y_test, y_pred)
rs = recall_score(y_test,y_pred)
f1 = f1_score(y_test,y_pred)
print(cm)
print()
print("Accuracy:")
print(acc)
print()
print("Precision:")
print(ps)
print()
print("Recall:")
print(rs)
print()
#F1 Score = 2* Precision Score * Recall Score/ (Precision Score + Recall Score/)
print("F1 Score:")
print (f1)
```
